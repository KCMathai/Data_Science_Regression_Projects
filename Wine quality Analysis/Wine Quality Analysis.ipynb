## Data Normalization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data = scaler.fit_transform(data)
```

```python
len(data[0])
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x=data[:,:-1]
y=data[:,-1]
```

```python
x_train, x_test, y_train, y_test = tts(x,y,test_size=0.1,random_state=69)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## Error and Accuracy Calculation

```python
from sklearn.metrics import mean_squared_error, r2_score

def metrics(y_true,y_pred):
    print("RSME:",mean_squared_error(y_true, y_pred)**0.5)
    print("R Squared Error:",r2_score(y_true, y_pred))
    
def accuracy(y_true, y_pred):
    errors = abs(y_true-y_pred)
    mean_absolute_percentage_error = 100 * np.mean(errors/y_true)
    accurate = 100 - mean_absolute_percentage_error
    return accurate
```

## Building a Model (Linear Regression)

```python
from sklearn.linear_model import LinearRegression

model= LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
metrics(y_test, y_pred)
```

```python
accuracy(y_test, y_pred)
```

## Building a Model ( Random Forest )

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor(n_estimators = 2000, min_samples_split = 3)
model_rf.fit(x_train, y_train)
```

```python
pred_rf = model_rf.predict(x_test)
```

```python
metrics(y_test, pred_rf)
```

```python
accuracy(y_test, pred_rf)
```

#### From the above 2 models, we can infer that Random Forest Model is better with 88.08 accuracy
